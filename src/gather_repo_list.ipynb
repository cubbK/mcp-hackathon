```python
from dotenv import load_dotenv
import os

load_dotenv()  # laddar .env-filen
GITHUB_TOKEN = os.getenv("GITHUB_TOKEN")

print(f"GITHUB_TOKEN: {GITHUB_TOKEN[:8] if GITHUB_TOKEN else 'Not found'}...")
```

```python
import requests

ORG = "hm-group"
PER_PAGE = 100  # max per page
repos = []
page = 1

# Om du har en GitHub token fr hgre rate limit:
HEADERS = {
    "Authorization": f"token {GITHUB_TOKEN}"
}

while True:
    try:
        # Hmta repos fr sidan
        url = f"https://api.github.com/orgs/{ORG}/repos?per_page={PER_PAGE}&page={page}"
        response = requests.get(url, headers=HEADERS)
        response.raise_for_status()
        
        data = response.json()
        if not data:
            break
        
        for repo in data:
            # Exempel: spara namn, url, privat/public
            repos.append({
                "name": repo["name"],
                "url": repo["html_url"],
                "private": repo["private"],
                "description": repo["description"]
            })
        
        print(f"Fetched page {page} ({len(data)} repos), {page*100} fetched")
        page += 1
    except requests.RequestException as e:
        print(f"Error fetching page {page}: {e}")
        break

repos
```

```python
# Spara till fil
import json
with open("all_repos.json", "w") as f:
    json.dump(repos, f, indent=2)

print(f"Done! Total repos fetched: {len(repos)}")
```
